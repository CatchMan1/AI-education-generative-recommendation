```python
import h5py
with h5py.File('emb/course_id_map.h5', 'r') as f1:
    course_id = f1['item_id'].asstr(encoding='utf-8')[:]
    course_num_id = f1['item_num_id'][:]
    course_id_map = dict(zip(course_id, course_num_id))
print(len(course_id_map))
```

```python
with h5py.File('emb/user_item_interact.h5', 'r') as f2:
    user_ids = f2['user_id'][:]          # 提取所有 user_id
    user_profile = f2['user_profile'].asstr(encoding='utf-8')[:] # 提取所有 user_profile
    item_lists = f2['item_id_list'][:]    # 提取所有变长数组
    rec_data = list(zip(user_ids, user_profile, item_lists))
print(rec_data[:5])
```

```python
with h5py.File('emb/course_item_embs.h5', 'r') as f:
    embeddings = f['item_embs'][:]
print(embeddings[0:20])
```

```python
with h5py.File('emb/user_item_emb.h5', 'r') as f2:
    user_ids = f2['user_id'][:]          # 提取所有 user_id
    item_lists = f2['user_embs'][:]    # 提取所有变长数组
    rec_data = list(zip(user_ids, item_lists))
print(len(rec_data))
```
